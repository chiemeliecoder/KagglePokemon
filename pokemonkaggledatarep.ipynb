```python
import time
full_pokemon_df_xl = pd.concat([complete_pokemon_original] * 500)

#The slow way
def sum(row):
    return row['attack'] + row['defense'] + row['speed']
    
start = time.time()
full_pokemon_df_xl['total_score_slow'] = full_pokemon_df_xl.apply(sum, axis=1)
end = time.time()
print(end-start)

# The fast way
start = time.time()
full_pokemon_df_xl['total_score_fast'] = full_pokemon_df_xl['attack'] + full_pokemon_df_xl['defense'] + full_pokemon_df_xl['speed']
end = time.time()
print(end-start)
```
